The values inside the training set and the set arrays are converted to float values as we are dividing them by 255 for normalizing in the next step

```
X_train = X_train.astype('float32') 
X_test = X_test.astype('float32')
```

```
X_train = X_train/255.0
X_test = X_test/255.0
```

In the code given below we are trying to get an optimal value of the number of clusters using the elbow method.
We find the wcss for a range of values and the value at which there is sharp change will be optimal number of clusters to used for training our model.
Here the value of K that we get is 220.

```
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.cluster import MiniBatchKMeans

wcss = []
for i in range(120,240,10):
    kmeans = MiniBatchKMeans(n_clusters = i, init = 'k-means++')
    kmeans.fit(X_train)
    wcss.append(kmeans.inertia_)
plt.plot(range(120,240,10), wcss)
plt.title('The Elbow Method')
plt.xlabel('Number of clusters')
plt.ylabel('WCSS')
plt.show()
```

We are now training our model using 220 clusters.The kmeans++ is responsible for performing random initialization
kmeans.labels_ contains all the cluster numbers

```
from sklearn.cluster import KMeans
from sklearn.cluster import MiniBatchKMeans
kmeans = KMeans(n_clusters = 220, init = 'k-means++')
kmeans.labels_ = kmeans.fit_predict(X_train)
```

i goes from 0 to len(np.unique(kmeans.labels_) which returns the total number of unique lables in kmeans.labels_.  
The index then contains an array of 1's and 0's wherein for every value belonging to the ith cluster value we get a 1 and elsewhere we get a 0.                                                                              
y_train[index==1] takes the number for every cluster where index is equal to 1 and finds y_train value at that paticular number.The number that occurs the most is the number that the images in that cluster belongs to.                                                                             reference_labels contains the value of the numbers for that cluster.                                    



```
def retrieve_info(cluster_labels,y_train):
  reference_labels={}
  for i in range(len(np.unique(kmeans.labels_))):
    index = np.where(cluster_labels == i,1,0)
    num = np.bincount(y_train[index==1]).argmax()
    reference_labels[i] = num
  return reference_labels
```

```
y_train=np.array(y_train)
y_train=y_train[:,0]
```

number_labels contains the value corresponding to the paticular image

```
reference_labels = retrieve_info(kmeans.labels_,y_train)
number_labels = np.random.rand(len(kmeans.labels_))
for i in range(len(kmeans.labels_)):
  number_labels[i] = reference_labels[kmeans.labels_[i]]
```

The accuracy is predicted by evaluating the y_train and the predicted labels

```
from sklearn.metrics import accuracy_score
print(accuracy_score(number_labels,y_train)*100)
```

The value for the images in the test set is then predicted and these values are stored in number_labels1

```
number_labels1=[]
a=kmeans.predict(X_test).tolist()
type(a)
for i in a:
  number_labels1.append(reference_labels[i])
```
